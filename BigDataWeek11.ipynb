```python
print("Average enrollment per department:")
avg_pipeline = [
    {'$group': {'_id': '$department', 'average_enrollment': {'$avg': '$enrollments'}}}
]

for result in courses_collection.aggregate(avg_pipeline):
    print(result)
```

```python
print("Maximum enrollment per department:")
max_pipeline = [
    {'$group': {'_id': '$department', 'max_enrollment': {'$max': '$enrollments'}}}
]

for result in courses_collection.aggregate(max_pipeline):
    print(result)
```

```python
pipeline = [
    {'$project': {'course_name': '$course', 'department_name': '$department', 'enrollments': 1}}
]

for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
pipeline = [
    {'$addFields': {'enrollment_category': {'$cond': {'if': {'$gt': ['$enrollments', 20]}, 'then': 'high', 'else': 'low'}}}}
]

for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
pipeline_HW1 = [
    {'$group': {'_id': '$department', 'total_courses': {'$sum': 1}}}
]

print("Count of courses per department:")
for result in courses_collection.aggregate(pipeline_HW1):
    print(result)
```

```python
pipeline_HW2 = [
    {'$match': {'enrollments': {'$gt': 25}, 'department': 'Computer Science'}},
    {'$group': {'_id': '$department', 'total_courses': {'$sum': 1}, 'total_enrollments': {'$sum': '$enrollments'}}}
]

print("Filtered and grouped result:")
for result in courses_collection.aggregate(pipeline_HW2):
    print(result)
```

```python
pipeline_HW2 = [
    {'$match': {'enrollments': {'$eq': 25}, 'department': 'Computer Science'}},
    {'$group': {'_id': '$department', 'total_courses': {'$sum': 1}, 'total_enrollments': {'$sum': '$enrollments'}}}
]

print("Filtered and grouped result:")
for result in courses_collection.aggregate(pipeline_HW2):
    print(result)
```

```python
students_collection = db['students']
students_collection.insert_many([
    {'student_name': 'Martha', 'course': 'Math 101'},
    {'student_name': 'Sam', 'course': 'CS 102'},
    {'student_name': 'Nika', 'course': 'CS 102'},
    {'student_name': 'Elenor', 'course': 'History 201'}
])

pipeline_HW3 = [
    {'$lookup': {'from': 'students', 'localField': 'course', 'foreignField': 'course', 'as': 'student_details'}},
    {'$project': {'_id': 0, 'course': 1, 'department': 1, 'enrollments': 1, 'students': 1, 'student_details.student_name': 1}}
]

print(" Joined data (courses + students):")
for result in courses_collection.aggregate(pipeline_HW3):
    print(result)
```
