```python
products = db['products']

products.delete_many({})

product_data = [
    {'name': 'Laptop Asus', 'category': 'Electronics', 'price': 12000000},
    {'name': 'Smartphone Samsung', 'category': 'Electronics', 'price': 9000000},
    {'name': 'Headphone Sony', 'category': 'Electronics', 'price': 2500000},
    {'name': 'Kemeja Batik', 'category': 'Fashion', 'price': 350000},
    {'name': 'Sepatu Nike', 'category': 'Fashion', 'price': 1200000},
    {'name': 'Meja Belajar', 'category': 'Furniture', 'price': 800000},
    {'name': 'Kursi Kantor', 'category': 'Furniture', 'price': 1500000},
    {'name': 'Kompor Gas', 'category': 'Home Appliance', 'price': 700000},
    {'name': 'Blender Philips', 'category': 'Home Appliance', 'price': 650000},
    {'name': 'Dispenser Aqua', 'category': 'Home Appliance', 'price': 950000}
]

products.insert_many(product_data)
print("10 data produk berhasil dimasukkan ke koleksi 'products'.")

for product in products.find():
    print(product)
```

```python
#NO4.2
```

```python
average_price_pipeline = [
    {'$group': {'_id': None, 'avgPrice': {'$avg': '$price'}}}
]

average_result = list(products.aggregate(average_price_pipeline))
average_price = average_result[0]['avgPrice']

print(f"Rata-rata harga produk: Rp{average_price:,.0f}")

print("\nProduk dengan harga di atas rata-rata:")
for product in products.find({'price': {'$gt': average_price}}):
    print(f"{product['name']} ({product['category']}) : Rp{product['price']:,}")
```

```python
pipeline = [
    {'$group': {'_id': '$category', 'total_products': {'$sum': 1}}}
]

print("Total produk per kategori:")
for result in products.aggregate(pipeline):
    print(f"{result['_id']}: {result['total_products']} produk")
```

```python
pipeline = [
    {'$sort': {'department': 1, 'salary': -1}}, 
    {'$group': {
        '_id': '$department',
        'top_employees': {
            '$push': {
                'name': '$name',
                'salary': '$salary',
                'age': '$age'
            }
        }
    }},
    {'$project': {
        '_id': 1,
        'top_employees': {'$slice': ['$top_employees', 5]}  
    }}
]

print("5 Karyawan dengan gaji tertinggi per departemen:")
for dept in collection.aggregate(pipeline):
    print(f"\nDepartemen: {dept['_id']}")
    for emp in dept['top_employees']:
        print(f"{emp['name']} | Gaji: {emp['salary']} | Umur: {emp['age']}")
```

```python
pipeline = [
    {
        '$group': {
            '_id': '$department',              
            'total_salary': {'$sum': '$salary'},  
            'avg_age': {'$avg': '$age'}           
        }
    },
    {
        '$sort': {'_id': 1} 
    }
]

print("Laporan Total Gaji & Rata-rata Umur per Departemen:\n")
for dept in collection.aggregate(pipeline):
    print(f"- Departemen: {dept['_id']}")
    print(f"  Total Gaji   : Rp{dept['total_salary']:,}")
    print(f"  Rata-rata Umur: {dept['avg_age']:.2f} tahun\n")
```
